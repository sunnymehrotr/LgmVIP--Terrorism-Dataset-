```python
df_1 = df.rename(columns= {'iyear':'year','imonth':'month', 'iday':'day','extended':'Extended','country_txt':'Country', 'provstate':'State','region_txt':'Region', 'city':'city','latitude':'latitude','longitude':'longitude','attacktype1_txt':'attacktype', 'nkill':'Killed', 'nwound':'wounded', 'target1':'Target','summary':'Summary', 'gname':'group','weaptype1_txt':'weapontype' ,'motive':'motive','targtype1_txt':'targettype'})
```

```python
df_1=df_1[['year','month','day','Extended','Country','State','Region','city','latitude','longitude','attacktype','Killed','wounded','Target','Summary','group','weapontype','motive','targettype']]
```

```python
df_1.head()
```

```python
df_1.isnull().sum()
```

```python
df_1.info()
```

```python
df_1.describe()
```

```python
print('maximum number of attacked in (year)',df_1['year'].value_counts().idxmax())
print('maximum number of attacked in (month)',df_1['month'].value_counts().idxmax())
print('maximum number of attacked in which region',df_1['Region'].value_counts().idxmax())
print('most type of attack ',df_1['attacktype'].value_counts().idxmax())
print('most attacked city is ',df_1['city'].value_counts().index[1])
```

```python
df_1.corr()
```

```python
plt.figure(figsize=(20,10))
sns.countplot(df_1['year'])
plt.title('terrorist attack each year')
```

```python
df_1['Killed']=df_1['Killed'].fillna(0)
df_1['wounded']=df_1['wounded'].fillna(0)
```

```python
df_1.head()
```

```python
df_1.Killed.isna().sum()
```

```python
df_2=df_1[['year','Killed']].groupby('year').sum().sort_values(by='Killed',ascending=False)
df_2.head()
```

```python
df_2.plot(kind='bar',figsize=(20,10))
plt.xlabel('year',fontsize=15)
plt.ylabel('Number of killed')
```

```python
df_3=df_1.Region.value_counts()
df_3
```

 Number of Attacks Region wise:

```python
df_3.plot(kind='bar')
plt.xlabel('Region')
plt.ylabel('Number of attacks')
```

```python
df_4=pd.crosstab(df_1['year'],df_1['Region'])
df_4.head()
```

```python
df_4.plot(kind='area',figsize=(15,8))
plt.xlabel('year')
plt.ylabel('Number of times')
plt.title('Region wise attack')
```

```python
df_5=df_1[['Region','Killed']].groupby('Region').sum().sort_values('Killed',ascending=False)
df_5
```

```python
df_5.plot(kind='bar')
plt.xlabel('Region')
plt.ylabel('Killed')
plt.title('Region wise people killed')
```
